# Análise de dados do UFC

O UFC (Ultimate Fighting Championship) é a maior organização de artes marciais mistas (MMA) do mundo. Fundado em 1993, ele promove combates em várias categorias de peso, onde lutadores de diferentes estilos, como jiu-jitsu, boxe, wrestling e kickboxing, competem em um octógono. O UFC é conhecido por suas lutas de alto nível e eventos globais.

## Objetivo

Este projeto busca esboçar as possibilidades que uma análise mais detalhada dos dados do UFC pode oferecer na tomada de decisão estratégica, auxiliando na identificação de padrões de desempenho, na previsão de resultados e na otimização de estratégias de combate.

## Coleta dos dados

Os dados foram obtivos através do Kaggle e foram coletados desde meados de 2010 até o UFC 300 em 13 de abril de 2024.

É importante identificar os tipos de dados para realizar o tratamento adequado na hora de fazermos o pré-processamento.

```python
t_colunas = df1.dtypes
n_colunas = df1.columns
i_object = []
i_float = []
i_int = []
i_bool = []
for i, item in enumerate(t_colunas):
    if item == object:
        i_object.append(n_colunas[i])
    elif item == 'float64':
        i_float.append(n_colunas[i])
    elif item == 'int64':
        i_int.append(n_colunas[i]) 
    else:
        i_bool.append(n_colunas[i])
```

```python
print("Colunas do tipo 'object'")
for i in i_object:
    print(f'-{i}')
```

## Visualização dos dados

```python
# utilizando o mapa de calor para determinar a correlação entre as variáveis
plt.figure(figsize=(16,9))

sns.heatmap(df1.corr(), annot=False, linewidths = 1, cmap="coolwarm", linecolor="k")
```

Devido ao grande número de colunas no dataframe, não é possível identificar facilmente quais variáveis estão correlacionadas entre si. Portanto, faremos um subset selecionando apenas as variáveis que apresentam uma correlação alta, com um nível de 0,75 ou mais.

```python
corr_matrix = df1.corr()
high_corr = corr_matrix[corr_matrix  > 0.75]
high_corr = high_corr.where((corr_matrix != 1))
high_corr_pairs = high_corr.stack().reset_index()
high_corr_pairs.head(20)
```

Algumas dessas colunas já esperavamos como as Odds e as expectitivas de valores, e a sequência de vitorias de um lutador e sua influencia no total de vitorias.
